# 2D-графики, гистограмы и ящик с усами

```python
import matplotlib.pyplot as plt
import numpy as np
# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)
# plot
plt.figure(figsize = (15,5))
plt.subplot(121)
result1 = plt.hist(x,bins=10,edgecolor='white', linewidth=2, align = 'left')
plt.subplot(1,2,2)
result2 = plt.hist(x,bins=8,edgecolor='white', linewidth=2)
plt.show()
print(f'Количество значений в диапазоне: {result1[0]}')
#print('Центры диапазонов:' + str(result[1]))#не рекомендуется
print(f'Границы диапазонов: {result1[1]}')
print(f'Количество значений в диапазоне: {result2[0]}')
#print('Центры диапазонов:' + str(result[1]))#не рекомендуется
print(f'Границы диапазонов: {result2[1]}')
```

# Построение трёхмерных графиков

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.cos(x)
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(x, y, z, label='parametric curve')
ax.view_init(45, 60)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 50)
y = np.arange(-np.pi, np.pi, 0.1)
X, Y = np.meshgrid(x,y)
# print(X)
# print(Y)
Z = np.cos(X) + np.cos(Y)
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z)
```
