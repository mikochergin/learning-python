## 1.2 Пример 1 (решение уравнения)
Модуль scipy integrate имеет две функции ode и odeint которые предназначены для решения систем обыкновенных дифференциальных уравнений (первого порядка с начальными условиями в одной точке (т.е. задача Коши)

Функция ode() более универсальная, а функция odeint() (ODE integrator) имеет более простой интерфейс и хорошо решает большинство задач

```from scipy integrate import odeint```
Функция odeint имеет три обязательных аргумента и много опций Она имеет следующий формат odeint
```odeint(func, y0, t[, args=(), ...])```

Допустим надо решить диф уравнение 1-го порядка
```y' = y(t)*t , y(2)=1```

````python
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
#Задаём функцию
def dydt(y, t):
    return -y*t
t = np.linspace(-2, 2, 51) # время
y0 = 1 # начальное значение
y = odeint(dydt, y0, t) # решаем ОДУ
y = np.array(y).flatten()
plt.plot(t, y, 'sr', linewidth=3) #построение графика
plt.show() # отображение графика
````

## 1.2 Пример 1 (решение уравнения)
Имеем систему уравнений

 \begin{gather*}
y'_1 = y_2\\
y'_2 = -y_2 - y_1\\
y_1(0) = 0\\
y_2(0) = 1
\end{gather*}

````python
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Запишем СУ в виде матричной функции
def f(y, t): 
    y1, y2 = y 
    return [y2, - y2 - y1]
t = np.linspace( 0, 10, 41) # вектор переменной "время"
y0 = [0, 1] # начальные значения
w = odeint(f, y0, t) # решение ОДУ
y1 = w[:,0] 
y2 = w[:,1] 
fig = plt.figure(facecolor='white') 
plt.plot(t, y1, '-o', t, y2, '-o', linewidth=2)
plt.legend(['y1', 'y2']) 
plt.ylabel("z") 
plt.xlabel("t") 
plt.grid(True) 
plt.show() # вывод графика
````
